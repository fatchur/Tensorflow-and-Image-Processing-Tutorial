```python
# import opencv
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

#### The most fundamental

1. The concept of image
2. Hello world in image processing: image grayscaling

```python
img = cv2.imread('face.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print img.shape
plt.imshow(img)
plt.show()
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray)
plt.show()
```

#### The next step
1. Image binarization techniques: mean vs gaussian vs otsu method
2. Image edge detection: Canny vs sobel
3. image smoothing/denoising: gaussian vs median

```python
# mean method
th_mean = cv2.adaptiveThreshold(img_gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
plt.imshow(th_mean)
plt.show()
```

```python
# gaussian method
th_gaussian = cv2.adaptiveThreshold(img_gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
plt.imshow(th_gaussian)
plt.show()
```

```python
# otsu method
ret2,th_otsu = cv2.threshold(img_gray, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.imshow(th_otsu)
plt.show()
```

||||||||||||||
Now Edge Detection
||||||||||||||

```python
# canny edge
canny = cv2.Canny(img_gray, 100, 200)
plt.imshow(canny)
plt.show()
```

```python
# sobel
sobelx = cv2.Sobel(img_gray,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(img_gray,cv2.CV_64F,0,1,ksize=5)
plt.imshow(sobelx)
plt.show()
plt.imshow(sobely)
plt.show()
```
